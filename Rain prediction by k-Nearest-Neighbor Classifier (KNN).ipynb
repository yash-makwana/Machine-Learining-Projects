```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
import numpy as np
```

```python
df = pd.read_csv('files/weather.csv', parse_dates=True, index_col=0)
df.head()
```

```python
df.dtypes
```

```python
data = df[['Humidity3pm','Pressure3pm','Cloud3pm','RainTomorrow']]
data.head()
```

```python
dataset = data.dropna()
```

```python
X =dataset[['Humidity3pm','Pressure3pm','Cloud3pm']]
y = dataset['RainTomorrow']
y = np.array([0 if value == 'No' else 1 for value in y])
```

```python
X_train,X_text,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
neigh = KNeighborsClassifier()
neigh.fit(X_train,y_train)
y_pred = neigh.predict(X_text)
accuracy_score(y_test,y_pred)
```
