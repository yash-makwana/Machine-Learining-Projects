```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/movies.csv')
df.head()
```

```python
df.info()
```

```python
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replace the null values with null string
for feature in selected_features:
  df[feature] = df[feature].fillna('')
```

```python
combined_features = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
combined_features
```

```python
vectorizer = TfidfVectorizer()
```

```python
feat_vector = vectorizer.fit_transform(combined_features)
```

```python
#Getting the simiilarity Score
similarity = cosine_similarity(feat_vector)
```

```python
movie_name = input('Enter your favourite movie name: ')
```

```python
#creating a list with all movie name given in the dataset
list_of_all_titles = df['title'].tolist()
print(list_of_all_titles)
```

```python
#Finding the close match for the movie name given by users
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
from operator import index
#Finding the index of the movie with title

index_of_the_movie = df[df.title == find_close_match[0]]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse = True)
print(sorted_similar_movies)
```

```python
print('Movies suggested for you: \n')

i = 0
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index==index]['title'].values[0]
  if (i<30):
    print(i,'.',title_from_index)
    i+=1
```
